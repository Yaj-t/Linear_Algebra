```python
def BackwardSubstitution(A, b):
    n = A.shape[0]
    x = np.zeros(n)

    for i in range(n-1, -1, -1):
        sum_ax = 0
        for j in range(i+1, n):
            sum_ax += A[i, j] * x[j]
        x[i] = (b[i] - sum_ax) / A[i, i]

    return x
```
